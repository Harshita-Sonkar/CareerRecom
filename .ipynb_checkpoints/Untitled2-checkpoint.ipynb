```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import NMF
from hmmlearn import hmm
from sklearn.cluster import KMeans
from gensim.models import Word2Vec
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
import random
import warnings
warnings.filterwarnings('ignore')
```

```python
occupation_data = pd.read_excel('ONET/Occupation Data.xlsx')
occupation_data.head()
```

```python
code_to_title = dict(zip(occupation_data['O*NET-SOC Code'], occupation_data['Title']))
skills_matrix = pd.pivot_table(
        skills_data,
        values='Data Value',
        index='O*NET-SOC Code',
        columns='Element Name',
        fill_value=0
)
skills_matrix.index = skills_matrix.index.map(lambda x: code_to_title.get(x, x))

skills_matrix.to_csv('skills_matrix.csv')
```

```python
skills_matrix
```

```python
occupation_similarity = pd.DataFrame(
    cosine_similarity(skills_matrix),
    index=skills_matrix.index,
    columns=skills_matrix.index
)
occupation_similarity.to_csv('occupation_similarity.csv')
occupation_similarity
```

```python
career_transitions = pd.DataFrame(0, index=skills_matrix.index, columns=skills_matrix.index)
for occupation in skills_matrix.index:
    similar_occupations = occupation_similarity[occupation].sort_values(ascending=False).index[1:6]  # Top 5 similar occupations
    career_transitions.loc[occupation, similar_occupations] = 1
career_transitions.to_csv('career_transitions.csv')
career_transitions
```

```python
def train_nmf_model(skills_matrix, n_components=10):
    model = NMF(n_components=n_components, init='random', random_state=42)
    W = model.fit_transform(skills_matrix)  # User-occupation matrix
    H = model.components_  # Occupation-feature matrix
    plt.figure(figsize=(12, 8))
    sns.heatmap(H, cmap='viridis', annot=True, fmt=".2f")
    plt.title("NMF Components (Occupation-Feature Matrix)")
    plt.xlabel("Features")
    plt.ylabel("Occupations")
    plt.show()
    
    return model, W, H
nmf_model, W, H = train_nmf_model(skills_matrix)
```

```python
def train_hmm_model(career_transitions, n_components=10):
    sequences = career_transitions.values.tolist()
    model = hmm.MultinomialHMM(n_components=n_components)
    model.fit(sequences)
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(model.transmat_, cmap='YlOrRd', annot=True, fmt=".2f")
    plt.title("HMM Transition Matrix")
    plt.xlabel("Next Occupation")
    plt.ylabel("Current Occupation")
    plt.show()
    
    return model

hmm_model = train_hmm_model(career_transitions)
```

```python
def train_kmeans_model(user_profiles, n_clusters=5):
    print("Training K-Means model...")
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    user_clusters = kmeans.fit_predict(user_profiles)
    
    plt.figure(figsize=(10, 8))
    sns.scatterplot(x=user_profiles[:, 0], y=user_profiles[:, 1], hue=user_clusters, palette='viridis', s=100)
    plt.title("User Clusters")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
    
    return kmeans, user_clusters
```

```python
def run_career_assessment_quiz():
    scenarios = generate_scenario_questions()
    responses = []
    
    print("=== Career Path Assessment Quiz ===")
    selected_scenarios = random.sample(scenarios, 5)
    
    for i, scenario in enumerate(selected_scenarios, 1):
        print(f"\nScenario {i}: {scenario['question']}")
        for j, option in enumerate(scenario['options'], 1):
            print(f"  {j}. {option['text']}")
        
        while True:
            try:
                choice = int(input("\nEnter your choice (1-4): "))
                if 1 <= choice <= 4:
                    selected_skills = scenario['options'][choice-1]['skills']
                    responses.extend(selected_skills)
                    break
                else:
                    print("Please enter a number between 1 and 4.")
            except ValueError:
                print("Please enter a valid number.")
    
    # Visualize user skills
    user_skills_count = pd.Series(responses).value_counts()
    plt.figure(figsize=(10, 6))
    sns.barplot(x=user_skills_count.index, y=user_skills_count.values, palette='viridis')
    plt.title("User Skills Profile")
    plt.xlabel("Skills")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.show()
    
    return responses
```

```python
def map_skills_to_careers(user_skills, nmf_model, W, H, top_n=3):
    print("Mapping skills to careers...")
    user_profile = pd.Series(0, index=skills_matrix.columns)
    for skill in user_skills:
        if skill in user_profile.index:
            user_profile[skill] += 1
    
    user_profile = user_profile.values.reshape(1, -1)
    user_embedding = nmf_model.transform(user_profile)
    predicted_scores = np.dot(user_embedding, H)
    
    top_occupations = skills_matrix.index[np.argsort(predicted_scores[0])[-top_n:]]
    return top_occupations
```

```python
def generate_career_paths(start_occupation, hmm_model, depth=2):
    print("Generating career paths...")
    path = [start_occupation]
    current_occupation = start_occupation
    
    for _ in range(depth):
        next_occupation = hmm_model.predict([current_occupation])[0]
        path.append(next_occupation)
        current_occupation = next_occupation
    
    return path
```

```python
def display_career_paths(career_paths, skills_matrix, education_data, occupation_data):
    print("\n=== Your Recommended Career Paths ===")
    for i, (start_occupation, path) in enumerate(career_paths.items(), 1):
        print(f"\nCareer Path {i} - Starting as: {start_occupation}")
        
        # Visualize the career path
        G = nx.DiGraph()
        for j in range(len(path) - 1):
            G.add_edge(path[j], path[j + 1])
        
        plt.figure(figsize=(10, 6))
        pos = nx.spring_layout(G, seed=42)
        nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', font_size=10, font_weight='bold', edge_color='gray')
        plt.title(f"Career Path {i}: {start_occupation}")
        plt.show()
        
        for j, occupation in enumerate(path):
            print(f"\n  {j+1}. {occupation}")
            occupation_skills = skills_matrix.loc[occupation].sort_values(ascending=False).head(5)
            print("     Key skills required:")
            for skill, value in occupation_skills.items():
                if value > 0:
                    print(f"     - {skill}")
            
            try:
                occupation_row = occupation_data[occupation_data['Title'] == occupation]
                if len(occupation_row) > 0:
                    soc_code = occupation_row['O*NET-SOC Code'].values[0]
                    education_info = education_data[education_data['O*NET-SOC Code'] == soc_code]
                    if len(education_info) > 0:
                        edu_categories = education_info['Category'].unique()
                        if len(edu_categories) > 0:
                            print(f"     Typical education: {edu_categories[0]}")
            except:
                pass
```

```python
# Cell 11: Define the end-to-end assessment and recommendation process
def career_path_assessment():
    print("Welcome to the Career Path Recommender!")
    print("This tool will help you discover potential career paths based on your preferences.")
    print("You'll answer a series of scenario-based questions, and we'll recommend suitable career paths.")
    
    input("\nPress Enter to begin the assessment...")
    
    # Run the quiz
    user_skills = run_career_assessment_quiz()
    
    print("\nAnalyzing your responses...")
    
    # Map skills to careers using NMF
    top_occupations = map_skills_to_careers(user_skills, nmf_model, W, H)
    
    print(f"\nBased on your responses, we've identified {len(top_occupations)} potential starting points:")
    for i, occupation in enumerate(top_occupations, 1):
        print(f"{i}. {occupation}")
    
    # Generate career paths from recommended occupations
    career_paths = {}
    for occupation in top_occupations:
        career_paths[occupation] = generate_career_paths(occupation, hmm_model)
    
    # Display detailed career paths with visualizations
    display_career_paths(career_paths, skills_matrix, education_data, occupation_data)
    
    print("\nThank you for using the Career Path Recommender!")
    
    # Optional: Visualize the top career path
    visualize_option = input("\nWould you like to visualize your top career path? (yes/no): ")
    if visualize_option.lower() == 'yes':
        top_occupation = top_occupations[0]
        visualize_career_path(top_occupation, career_paths[top_occupation])
```
